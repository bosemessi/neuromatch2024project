```python
datafolder = "D:\\neuromatch2024\\datasets\\project\\Allen\\temp\\combined_files\\"

i = 1

ophys_session_id = gdf.session_id[i]
mouse = gdf.mouse_id[i]
experience = gdf.experience_level[i]
cre_line = "vip" if gdf.cre_line[i] == 'Vip-IRES-Cre' else "sst"
dominance = gdf.dominance[i]

print(ophys_session_id, mouse, experience)

with open(f'{datafolder}masterfile_{ophys_session_id}.pickle', 'rb') as handle:
    masterdict = pickle.load(handle)

neural_data = masterdict["neural_data"]
stimulus_table = masterdict["stimulus_table"]
trials = masterdict["trials"]
licks = masterdict["licks"]
rewards = masterdict["rewards"]
eyetracking = masterdict["eyetracking"]
running = masterdict["running"]
```

```python
licks["dT"] = licks.timestamps.diff().fillna(0)
lickstarts = licks[licks.dT >= 1].timestamps.values

#### misses

misses = []

for start_time in stimulus_table[
    (stimulus_table.is_change)
    & (stimulus_table.active)
    & (stimulus_table.start_time.between(300, 1000))
].start_time.unique():
    anylick = [l for l in lickstarts if (l >= start_time) & (l <= start_time + 0.75)]
    if len(anylick) == 0:
        misses.append(start_time)
misses = np.array(misses)

#### hits

hits = []
for lickstart in lickstarts:
    # print(lickstart)
    stm = stimulus_table[
        (stimulus_table.start_time.between(lickstart - 0.75, lickstart))
        & (stimulus_table.is_change)
    ]
    # print(lickstart, len(stm))

    if len(stm) > 0:
        # hits.append(stm.start_time.item())
        hits.append(lickstart)
hits = np.array(hits)
```

```python
unique_images = stimulus_table[~stimulus_table.omitted].image_name.unique()
img_to_color = dict(zip(unique_images, ["#66c2a5","#fc8d62","#8da0cb","#e78ac3","#a6d854","#ffd92f","#e5c494","#b3b3b3"]))
img_to_color
```

```python
fig, ax = plt.subplots(figsize=(15, 10), nrows=3, sharex=True)

t1 = 500
t2 = 1000

for _, stimulus in stimulus_table[(stimulus_table.start_time.between(t1, t2))].iterrows():
    if stimulus["image_name"] == "omitted":
        continue 
    else:
        ax[0].axvspan(stimulus.start_time, stimulus.end_time, color=img_to_color[stimulus["image_name"]])
ax[0].set_title("Image presentations");

ax[1].eventplot(
    hits[(hits >= t1) & (hits <= t2)],
    color="#44AA99",
    label="Hits",
    zorder=5,
)
ax[1].eventplot(
    misses[(misses >= t1) & (misses <= t2)],
    color="#882255",
    ls="--",
    label="Misses",
    zorder=5,
)
ax[1].eventplot(
    licks[licks.timestamps.between(t1, t2)].timestamps,
    color = "grey",
    alpha = 0.1,
    label = "All licks",
    zorder = 2
)
ax[1].legend()
ax[1].set_title("Licks")

ax[2].eventplot(
    rewards[rewards.timestamps.between(t1, t2)].timestamps,
    color = "tab:blue",
    zorder = 2
)
ax[2].set_title("Rewards")

ax[2].set_xlabel("Timestamp in seconds")

for axes in ax:
    axes.set_yticks([])
```

```python
cells = neural_data.cell_specimen_id.unique()

hitdata = [pd.DataFrame()] * len(cells)
missdata = [pd.DataFrame()] * len(cells)

for i, cell_id in enumerate(cells):
    df_hit = utilities.event_triggered_response(
        data=neural_data.query('cell_specimen_id == @cell_id'),
        t='timestamps',
        y='dff',
        event_times = hits, #stimulus_table[stimulus_table.omitted]['start_time'],
        t_before=3,
        t_after=3,
        output_sampling_rate=50,
    )
    df_miss = utilities.event_triggered_response(
        data=neural_data.query('cell_specimen_id == @cell_id'),
        t='timestamps',
        y='dff',
        event_times = misses, #stimulus_table[stimulus_table.omitted]['start_time'],
        t_before=3,
        t_after=3,
        output_sampling_rate=50,
    )
    df_hit["cell_id"] = cell_id
    df_miss["cell_id"] = cell_id

    hitdata[i] = df_hit
    missdata[i] = df_miss

hitdata = pd.concat(hitdata).reset_index(drop=True)
missdata = pd.concat(missdata).reset_index(drop=True)
```

```python
hitdata.head()
```

```python
hit_traces = []
for cell_id in cells:
    hit_cell = hitdata[hitdata.cell_id == cell_id]
    for event_time in hit_cell.event_time.unique():
        hit_traces.append(hit_cell[hit_cell.event_time == event_time].dff.values)
hit_traces = np.array(hit_traces)
hit_traces.shape
```

```python
miss_traces = []
for cell_id in cells:
    miss_cell = missdata[missdata.cell_id == cell_id]
    for event_time in miss_cell.event_time.unique():
        miss_traces.append(miss_cell[miss_cell.event_time == event_time].dff.values)
miss_traces = np.array(miss_traces)
miss_traces.shape
```

```python
fig,ax = plt.subplots(figsize = (5,5))
ax.plot(np.arange(-3, 3.02, 0.02),  hit_traces.mean(axis = 0))
ax.plot(np.arange(-3, 3.02, 0.02),  miss_traces.mean(axis = 0))
```

```python
datadict = {
    "hit_trace_mean" : list(hit_traces.mean(axis = 0)),
    "miss_trace_mean" : list(miss_traces.mean(axis = 0)),
}
```
