```python
def plotter(i):
    ophys_session_id = gdf.session_id[i]
    mouse = gdf.mouse_id[i]
    experience = gdf.experience_level[i]
    cre_line = "vip" if gdf.cre_line[i] == 'Vip-IRES-Cre' else "sst"
    dominance = gdf.dominance[i]
    
    print(ophys_session_id, mouse, experience)
    
    with open(f'{datafolder}masterfile_{ophys_session_id}.pickle', 'rb') as handle:
        masterdict = pickle.load(handle)
    
    neural_data = masterdict["neural_data"]
    stimulus_table = masterdict["stimulus_table"]
    trials = masterdict["trials"]
    licks = masterdict["licks"]
    rewards = masterdict["rewards"]
    eyetracking = masterdict["eyetracking"]
    running = masterdict["running"]
    licks_annotated = annotate_licks(licks, rewards, stimulus_table)
    stimulus_table_annotated = annotate_bouts(licks_annotated, stimulus_table)
    duration = []
    latency = []
    
    for bout_number in licks_annotated.bout_number.unique():
        bout_start = licks_annotated[(licks_annotated.bout_number == bout_number) & (licks_annotated.bout_start)].timestamps.item()
        bout_end = licks_annotated[(licks_annotated.bout_number == bout_number) & (licks_annotated.bout_end)].timestamps.item()
        duration.append(bout_end - bout_start)
        stimulus_time = stimulus_table_annotated[stimulus_table_annotated.bout_number == bout_number].start_time.min()
        latency.append(bout_start - stimulus_time)

    fig,ax = plt.subplots(nrows=1, ncols=2, figsize=(15, 5))

    ax[0].plot(gaussian_filter1d(duration, 3))
    ax[1].plot(gaussian_filter1d(latency, 3))

    ax[0].set_xlabel("Bout number")
    ax[1].set_xlabel("Bout number")

    ax[0].set_ylabel("Duration")
    ax[1].set_ylabel("Latency")
```
